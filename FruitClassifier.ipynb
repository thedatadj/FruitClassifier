```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# I loaded the data

fruit_data = pd.read_table("/content/drive/MyDrive/Colab Notebooks/Workspace/Machine Learning/fruit_data_with_colors.txt")
```

```python
fruit_data.shape
```

There are 59 examples, in this dataset, and seven columns.

```python
fruit_data.head()
```

```python
# I split the data into two datasets: X and y

X = fruit_data[["mass", "width", "height", "color_score"]]
y = fruit_data["fruit_label"]


# Store the fruit names in a separate variable
target_class_names = fruit_data.fruit_name.unique()
```

```python
target_class_names
```

```python
# I split the datasets into: training and testing sets.

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


# This shows me the distribution of the data
shapes = {"X": [X.shape, X_train.shape, X_test.shape],
          "y": [y.shape, y_train.shape, y_test.shape],
          "Percent": ["100%", "75%", "25%"]}

pd.DataFrame(shapes, index=["Data", "Train", "Test"])
```

```python
# I trained a KNN classifier

from sklearn.neighbors import KNeighborsClassifier


model = KNeighborsClassifier(n_neighbors=5)

model.fit(X_train, y_train)
```

```python
# I want to see the model's performance on the training and test sets

train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("Train score", train_score, "\nTest score", test_score)
```

This is a very bad performance 😞, the model is underfitting:

```python
# Here I tried to scaled the features to improve the performance

from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()

Xscaled_train = scaler.fit_transform(X_train)

Xscaled_test = scaler.transform(X_test)
```

```python
# Here I compare the feature before and after scaling

print(Xscaled_train[:5],"\n", "\n", X_train.head())
```

```python
# I trained the classifier with the scaled features

model.fit(Xscaled_train, y_train)


# and see its performance on the training and test scaled sets
Scaled_train_score = model.score(Xscaled_train, y_train)
Scaled_test_score = model.score(Xscaled_test, y_test)

print("Scaled Train score", Scaled_train_score, "\nScaled Test score", Scaled_test_score)
```

The performance is way better now. 🙂

```python
# This is a summary of the results

results_data = {"Normal": [train_score, test_score],
           "Scaled": [Scaled_train_score, Scaled_test_score]}

results = pd.DataFrame(results_data, index=["Train", "Test"])
results.columns.name = "Performance"

results
```
